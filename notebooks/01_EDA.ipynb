```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

fig_dir = Path("../output/figures")
fig_dir.mkdir(parents=True, exist_ok=True)

# =========================
# 1. Đọc dữ liệu
# =========================

file_path = "../data/processed/features_final.csv"
df = pd.read_csv(file_path)

print("Số lượng mẫu:", len(df))
print("\nCác cột trong dữ liệu:")
print(df.columns.tolist())

# Đổi tên cột label nếu cần
label_col = "label"

print("\nPhân phối label:")
print(df[label_col].value_counts())

print("\nTỷ lệ label:")
print(df[label_col].value_counts(normalize=True) * 100)

# Bar chart slop vs genuine
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x=label_col, palette="Set2")
plt.title("Phân phối số lượng Slop vs Genuine")
plt.xlabel("Label")
plt.ylabel("Số lượng mẫu")
plt.tight_layout()
plt.savefig(fig_dir / "01_label_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```
